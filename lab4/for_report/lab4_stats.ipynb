```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# Вспомогательная функция для обработки ошибок и получения моды
def safe_mode(series):
    """Безопасное вычисление моды для pandas Series, возвращает None, если мода не найдена."""
    try:
        return Counter(series).most_common(1)[0][0]
    except IndexError:  # Если series пустая или все значения уникальны
        return None

def analyze_csv_files(folder_path):
    """
    Анализирует CSV файлы в указанной папке и выводит количество вхождений текстом
    для столбцов 'Symptoms', 'Analysis', 'Doctor' и строит гистограмму для 'Price'.

    Args:
        folder_path (str): Путь к папке с CSV файлами.
    """

    # Получаем список CSV файлов и сортируем по дате создания.
    filepaths = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith(".csv")]
    filepaths.sort(key=os.path.getctime)  # Сортировка по дате создания

    for filepath in filepaths:
        filename = os.path.basename(filepath)
        print(f"Обработка файла: {filename}")

        try:
            df = pd.read_csv(filepath)

            # Выводим количество вхождений для категориальных столбцов текстом
            print(f"Количество вхождений для Symptoms:")
            try:
                for value, count in df['Symptoms'].value_counts().items():
                    print(f"  {value}: {count}")
            except KeyError:
                print("  Столбец 'Symptoms' отсутствует.")

            print(f"\nКоличество вхождений для Analysis:")
            try:
                for value, count in df['Analysis'].value_counts().items():
                    print(f"  {value}: {count}")
            except KeyError:
                print("  Столбец 'Analysis' отсутствует.")

            print(f"\nКоличество вхождений для Doctor:")
            try:
                for value, count in df['Doctor'].value_counts().items():
                    print(f"  {value}: {count}")
            except KeyError:
                print("  Столбец 'Doctor' отсутствует.")

            # Строим гистограмму для 'Price'
            try:
                plt.figure(figsize=(8, 6)) # Создаем новую фигуру для гистограммы
                sns.histplot(df['Price'], kde=True)  # Используем histplot для гистограммы
                plt.title(f"Распределение Price для файла: {filename}")
                plt.xlabel("Price")
                plt.ylabel("Частота")
                plt.show() # Показываем график

            except KeyError:
                print("  Столбец 'Price' отсутствует.")
            except TypeError:
                print("  Столбец 'Price' содержит нечисловые данные. Невозможно построить гистограмму.")
            except Exception as e:
                print(f"Ошибка при построении графика для 'Price' в {filename}: {e}")


            print("-" * 40)

        except FileNotFoundError:
            print(f"Файл не найден: {filepath}")
        except pd.errors.EmptyDataError:
            print(f"Файл пуст: {filepath}")
        except pd.errors.ParserError:
            print(f"Ошибка парсинга файла: {filepath}.  Возможно, файл поврежден или имеет неправильный формат CSV.")
        except Exception as e:
            print(f"Непредвиденная ошибка при обработке файла {filename}: {e}")


# Замените на фактический путь к папке
folder_path = "C://Users//Zver//Desktop//uni//ALG//lab4//for_report"
analyze_csv_files(folder_path)
```
